```python
import pandas as pd
import matplotlib.pyplot as plt

mydata = pd.read_csv("data.csv")
mydata.fillna(100)
```

```python
mydata.columns
```

# Histogram

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-bright')

%matplotlib notebook

hist1 = mydata['dept_1 sales'].plot.hist(bins=20, figsize=(8,4), color='pink', ec="black", grid=True, alpha=0.95, legend=True)

hist1.grid(axis='y', alpha=0.6)
hist1.set_xlabel("Dept_1 Sales values")
hist1.set_title("Histogram- Distribution of Department_1_Sales")

plt.show()

hist1.figure.savefig("Histogram1.png")
```

# Histogram with mutiple variables in a single frame

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-bright')

%matplotlib notebook

hist2 = mydata[['dept_1 sales','dept_2 sales']].plot.hist(bins=20, figsize=(8,4), color=['pink', 'silver'], ec="black", grid=True, alpha=0.65, legend=True)

hist2.grid(axis='y', alpha=0.6)
hist2.set_xlabel("Dept_1,2 Sales values")
hist2.set_title("Histogram- Distribution of Department_1,2 Sales")

plt.show()

hist2.figure.savefig("Histogram2.png")
```

# Histogram with multiple variables in different frames

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('default')

%matplotlib notebook

hist3 = mydata[['dept_1 sales','dept_2 sales','dept_3 sales']].hist(bins=20, figsize=(10,4), color="lavender", ec="black", grid=True)


plt.show()
```

# BarChart

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('classic')

%matplotlib notebook

barch = mydata.plot.bar(x='Day type', y='Temperature', color="red", figsize=(20,5), legend=True)
barch.figure.savefig("barchart.png") 
```

# LineChart

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('dark_background')

%matplotlib notebook

line = mydata.plot.line(x='Date', y='CPI', color='lightblue', figsize=(12,6), legend=True)

line.figure.savefig("linechart.png")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('dark_background')

%matplotlib notebook

line2 = mydata.plot('Date', y= ['dept_1 sales','dept_2 sales','dept_3 sales'], figsize=(12,6))
line2.figure.savefig("linechart2.png")
```

# AreaChart

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')

%matplotlib notebook

area = mydata.plot.area(x='Date', y= ['Fuel_Price', 'Temperature'], color=['green', 'red','orange'], figsize=(10,4), alpha=0.5, stacked=False)

area.figure.savefig("AreaChart.png")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('classic')

%matplotlib notebook

area1 = mydata.plot.area(x='Date', figsize=(10,4), alpha=0.5, stacked=False)
plt.legend()
area1.figure.savefig("areachart1.png")
```

# ScatterPlot

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')

%matplotlib notebook


scatter = mydata.plot.scatter(x='Temperature', y='CPI', c=mydata.CPI, cmap='plasma', ec="black")
scatter.axvline(65, c=(.5,.5,.5), ls='--')
scatter.axhline(217, c=(.5,.5,.5), ls='--')
plt.show()

scatter.figure.savefig("scatterplot.png") 
```
